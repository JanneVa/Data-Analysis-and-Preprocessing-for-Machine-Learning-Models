3) We evaluate if our data frame have multiple variables stored in a column.


```python
def check_delimiters_in_columns(df, delimiters=[",", ";", "|", "-", "_"]):
    columns_with_delimiter = {}

    for col in df.columns:
        for delim in delimiters:
            contains_delimiter = df[col].astype(str).str.contains(delim, regex=False).sum()
            if contains_delimiter > 0:
                if col in columns_with_delimiter:
                    columns_with_delimiter[col][delim] = contains_delimiter
                else:
                    columns_with_delimiter[col] = {delim: contains_delimiter}

    return columns_with_delimiter
```

We call the function to evaluate if we have delimiters per column

```python
Entidad_delimiter_counts = check_delimiters_in_columns(Entidad)
print(Entidad_delimiter_counts)
```

5) Multiple types of units stored in the same tables
In this case, according to the indicator descriptor, it is shown that the counts are considered per person, housing and households, which is why units of measurement such as Kg or grams are not considered. However,
To address this point we consider the approach of analyzing descriptive statistics (mean, median, standard deviation, minimum, maximum) that may offer clues. For example, if a column expected to be in meters has average values in the range of thousands or millions, it can be suspected to be in millimeters or micrometers. Likewise, an analysis was carried out to evaluate possible inconsistencies in the scale by column.

```python
print(Entidad.describe())
```

```python
def check_scale_inconsistency(data):
    # Seleccionamos solo columnas numÃ©ricas
    numeric_columns = data.select_dtypes(include=['float64', 'int64'])

    inconsistent_columns = []

    for col in numeric_columns.columns:
        min_value = numeric_columns[col].min()
        max_value = numeric_columns[col].max()

        # Si el valor mÃ¡ximo es al menos 1000 veces mayor que el mÃ­nimo (excluyendo ceros), puede haber una inconsistencia.
        if min_value > 0 and max_value / min_value >= 1000:
            inconsistent_columns.append(col)

    return inconsistent_columns

Entidad = pd.read_csv('conjunto_de_datos\INE_ENTIDAD_2020.CSV', encoding='latin1')
inconsistent_cols = check_scale_inconsistency(Entidad)

if inconsistent_cols:
    print("Columnas con posible inconsistencia en escala:", inconsistent_cols)
else:
    print("No se encontraron columnas con inconsistencia en escala.")
```

In this case the results can be explained because the columns evaluate people who speak an indigenous language.

```python
# Identificamos las columnas de tipo 'object'
columnas_obj = Distrito.select_dtypes(include='object').columns

# Imprimimos las columnas de tipo 'object'
print(columnas_obj)
```

2) Beging to standarize values

```python
frecuencia_valores = Distrito['INDIGENA'].value_counts()
print(frecuencia_valores)
```

```python
# Reemplazamos valores en la columna 'INDIGENA'
Distrito['INDIGENA'] = Distrito['INDIGENA'].replace({'NO': 0, 'SI': 1})

# Verificamos el resultado
print(Distrito['INDIGENA'])
```

```python
frecuencia_valores = Distrito['COMPLEJIDA'].value_counts()
print(frecuencia_valores)
```

```python
dummies = pd.get_dummies(Distrito['COMPLEJIDA'], prefix='')

# Renombramos las columnas para que coincidan con el formato que deseas
rename_dict = {
    '_Disperso 1': 'DIS1',
    '_Disperso 2': 'DIS2',
    '_Concentrado 1': 'CON1',
    '_Concentrado 2': 'CON2',
    '_Altamente Concentrado 1': 'ALT_CON_1',
    '_Altamente Concentrado 2': 'ALT_CON_2',
    '_ConcentraciÃ³n Media': 'CON_MED',
    '_Muy Disperso 1': 'M_DIS1',
    '_Muy Disperso 2': 'M_DIS2'
}

dummies.rename(columns=rename_dict, inplace=True)

# Concatenamos el DataFrame original con las columnas dummy
Distrito_dummies = pd.concat([Distrito, dummies], axis=1)

# Eliminamos la columna original
Distrito_dummies.drop('COMPLEJIDA', axis=1, inplace=True)

print(Distrito_dummies.head())
```

```python
# Columnas dummy que creamos
dummy_columns = [
    'ALT_CON_1',
    'ALT_CON_2',
    'CON_MED',
    'CON1',
    'CON2',
    'DIS1',
    'DIS2',
    'M_DIS1',
    'M_DIS2',
]

# Convertimos los valores True/False a 1/0 para las columnas dummy
for col in dummy_columns:
    Distrito_dummies[col] = Distrito_dummies[col].astype(int)

print(Distrito_dummies.head())
```
